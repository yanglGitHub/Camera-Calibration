```python
#!/usr/bin/env python
# coding: utf-8

import cv2
import glob
import os
import argparse
import matplotlib.pyplot as plt
import numpy as np
from CalibrationClass import CameraCalibration, StereoCalibration
from load_write_PMLAB_calibration import write_to_PMLAB
```

```python
imagepath = r'G:\2019122011275252\*.bmp'
OneCamera = CameraCalibration(imagepath, 4)
TwoCamera = StereoCalibration(imagepath, 4)
```

```python
OneCamera.calibrate_images()
TwoCamera.calibrate_images()
```

```python
TwoCamera.camera_model
```

```python
imagepoints_l = np.array(TwoCamera.imgpoints_l)
imagepoints_r = np.array(TwoCamera.imgpoints_r)
objpoints = np.array(TwoCamera.objpoints)
M = OneCamera.camera_model['Intrinsic']
d = OneCamera.camera_model['distortion']
img_shape = (2048, 2448)
```

```python
flags = 0
flags |= cv2.CALIB_FIX_INTRINSIC
# flags |= cv2.CALIB_FIX_PRINCIPAL_POINT
# flags |= cv2.CALIB_USE_INTRINSIC_GUESS
# flags |= cv2.CALIB_FIX_FOCAL_LENGTH
# flags |= cv2.CALIB_FIX_ASPECT_RATIO
# flags |= cv2.CALIB_ZERO_TANGENT_DIST
# flags |= cv2.CALIB_RATIONAL_MODEL
# flags |= cv2.CALIB_SAME_FOCAL_LENGTH
# flags |= cv2.CALIB_FIX_K3
# flags |= cv2.CALIB_FIX_K4
# flags |= cv2.CALIB_FIX_K5

stereocalib_criteria = (cv2.TERM_CRITERIA_MAX_ITER +
                        cv2.TERM_CRITERIA_EPS, 100, 1e-5)

ret, M1, d1, M2, d2, R, T, E, F = cv2.stereoCalibrate(objpoints, imagepoints_l, imagepoints_r,
                                                      M, d, M, d, img_shape,
                                                     flags = flags, criteria = stereocalib_criteria)

camera_model = dict([('ret', ret), ('M1', M1), ('M2', M2), ('dist1', d1), ('dist2', d2), ('R', R), ('T', T)])
write_to_PMLAB(camera_model, 'fix_intrisic.clbpj')
write_to_PMLAB(TwoCamera.camera_model, 'freedom_intrisic.clbpj')
```
